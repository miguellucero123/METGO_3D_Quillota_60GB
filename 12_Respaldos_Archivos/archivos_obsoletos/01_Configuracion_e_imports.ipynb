```python
# =============================================================================
# ‚öôÔ∏è CONFIGURACI√ìN E IMPORTS MEJORADOS - METGO 3D OPERATIVO
# Archivo: 01_Configuracion_e_Imports.ipynb
# Versi√≥n: 2.0 | Fecha: 2025-01-02
# Sistema Meteorol√≥gico Agr√≠cola Quillota - Versi√≥n Operativa
# =============================================================================

import sys
import os
import warnings
import yaml
from pathlib import Path
from datetime import datetime, timedelta, date
import logging

print("‚öôÔ∏è METGO 3D OPERATIVO - Configuraci√≥n e Imports Mejorados")
print("üìÖ Versi√≥n 2.0 - Sistema Meteorol√≥gico Agr√≠cola Quillota")
print("üîß Todas las mejoras implementadas")
print("=" * 70)

# =============================================================================
# CONFIGURACI√ìN DE ENTORNO Y WARNINGS MEJORADA
# =============================================================================

# Supresi√≥n inteligente de warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', message='.*ScriptRunContext.*')
warnings.filterwarnings('ignore', message='.*missing ScriptRunContext.*')
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Variables de entorno optimizadas
os.environ['STREAMLIT_BROWSER_GATHER_USAGE_STATS'] = 'false'
os.environ['PYTHONWARNINGS'] = 'ignore'
os.environ['OMP_NUM_THREADS'] = '4'  # Optimizaci√≥n para ML

print("‚úÖ Warnings suprimidos correctamente")
print("‚úÖ Variables de entorno optimizadas")

# =============================================================================
# CARGA DE CONFIGURACI√ìN CENTRALIZADA
# =============================================================================

def cargar_configuracion():
    """
    Cargar configuraci√≥n centralizada desde YAML
    """
    try:
        config_path = Path("config/config.yaml")
        if config_path.exists():
            with open(config_path, 'r', encoding='utf-8') as f:
                config = yaml.safe_load(f)
            print("‚úÖ Configuraci√≥n centralizada cargada")
            return config
        else:
            print("‚ö†Ô∏è Archivo config.yaml no encontrado, usando configuraci√≥n por defecto")
            return crear_configuracion_default()
    except Exception as e:
        print(f"‚ùå Error cargando configuraci√≥n: {e}")
        return crear_configuracion_default()

def crear_configuracion_default():
    """
    Crear configuraci√≥n por defecto si no existe archivo
    """
    return {
        'QUILLOTA': {
            'nombre': 'Quillota',
            'region': 'Valpara√≠so',
            'coordenadas': {'latitud': -32.8833, 'longitud': -71.25}
        },
        'METEOROLOGIA': {
            'umbrales': {
                'temperatura': {'helada_severa': -2.0, 'calor_extremo': 35.0},
                'precipitacion': {'lluvia_intensa': 20.0},
                'viento': {'fuerte': 25.0}
            }
        }
    }

# Cargar configuraci√≥n
CONFIG = cargar_configuracion()

# =============================================================================
# IMPORTS CIENT√çFICOS Y AN√ÅLISIS MEJORADOS
# =============================================================================

# Manipulaci√≥n de datos
import pandas as pd
import numpy as np
import json
import pickle

# Configurar pandas para mejor visualizaci√≥n
pd.set_option('display.max_columns', 25)
pd.set_option('display.width', 1200)
pd.set_option('display.precision', 2)
pd.set_option('display.max_rows', 100)

print("‚úÖ Pandas configurado para visualizaci√≥n optimizada")

# Visualizaci√≥n
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Configurar matplotlib
plt.style.use('default')
plt.rcParams.update({
    'figure.figsize': (12, 8),
    'font.size': 11,
    'axes.grid': True,
    'grid.alpha': 0.3
})

# Configurar seaborn
sns.set_palette("husl")
sns.set_style("whitegrid")

print("‚úÖ Librer√≠as de visualizaci√≥n configuradas")

# Machine Learning
try:
    from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
    from sklearn.linear_model import LinearRegression, Ridge, Lasso
    from sklearn.svm import SVR
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.preprocessing import StandardScaler, MinMaxScaler
    from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    from sklearn.pipeline import Pipeline
    import joblib
    print("‚úÖ Scikit-learn y dependencias ML cargadas")
except ImportError as e:
    print(f"‚ö†Ô∏è Error cargando ML: {e}")

# APIs y conectividad
try:
    import requests
    from requests.adapters import HTTPAdapter
    from urllib3.util.retry import Retry
    print("‚úÖ Requests y conectividad configurados")
except ImportError as e:
    print(f"‚ö†Ô∏è Error cargando requests: {e}")

# Utilidades adicionales
from datetime import datetime, timedelta
import time
import warnings
from typing import Dict, List, Tuple, Optional, Any

print("‚úÖ Utilidades adicionales cargadas")

# =============================================================================
# CONFIGURACI√ìN DE LOGGING MEJORADA
# =============================================================================

def configurar_logging():
    """
    Configurar sistema de logging estructurado
    """
    try:
        # Crear directorio de logs
        logs_dir = Path("logs")
        logs_dir.mkdir(exist_ok=True)
        
        # Configurar logging
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(logs_dir / 'metgo_operativo.log'),
                logging.StreamHandler()
            ]
        )
        
        logger = logging.getLogger("METGO_3D")
        logger.info("Sistema de logging inicializado")
        print("‚úÖ Sistema de logging configurado")
        return logger
    except Exception as e:
        print(f"‚ö†Ô∏è Error configurando logging: {e}")
        return None

# Inicializar logging
logger = configurar_logging()

# =============================================================================
# CONFIGURACI√ìN DE QUILLOTA MEJORADA
# =============================================================================

QUILLOTA_CONFIG = CONFIG.get('QUILLOTA', {
    'nombre': 'Quillota',
    'region': 'Valpara√≠so',
    'pais': 'Chile',
    'coordenadas': {'latitud': -32.8833, 'longitud': -71.25},
    'elevacion': 120,
    'poblacion': 97572,
    'superficie_agricola': 15000
})

UMBRALES_CRITICOS = CONFIG.get('METEOROLOGIA', {}).get('umbrales', {
    'temperatura': {
        'helada_severa': -2.0,
        'helada_moderada': 0.0,
        'calor_extremo': 35.0,
        'calor_moderado': 30.0
    },
    'precipitacion': {
        'lluvia_intensa': 20.0,
        'lluvia_moderada': 10.0
    },
    'viento': {
        'fuerte': 25.0,
        'moderado': 15.0
    },
    'humedad': {
        'muy_baja': 30.0,
        'muy_alta': 85.0
    }
})

print(f"üìç Ubicaci√≥n: {QUILLOTA_CONFIG['nombre']}, {QUILLOTA_CONFIG['region']}")
print(f"üó∫Ô∏è Coordenadas: {QUILLOTA_CONFIG['coordenadas']['latitud']}, {QUILLOTA_CONFIG['coordenadas']['longitud']}")
print(f"üå°Ô∏è Umbrales cr√≠ticos configurados: {len(UMBRALES_CRITICOS)} categor√≠as")

# =============================================================================
# FUNCIONES DE UTILIDAD MEJORADAS
# =============================================================================

def verificar_dependencias():
    """
    Verificar que todas las dependencias cr√≠ticas est√©n disponibles
    """
    dependencias_criticas = [
        'pandas', 'numpy', 'matplotlib', 'seaborn', 
        'sklearn', 'requests', 'plotly'
    ]
    
    faltantes = []
    for dep in dependencias_criticas:
        try:
            __import__(dep)
        except ImportError:
            faltantes.append(dep)
    
    if faltantes:
        print(f"‚ö†Ô∏è Dependencias faltantes: {faltantes}")
        print("üí° Ejecutar: pip install -r requirements.txt")
        return False
    else:
        print("‚úÖ Todas las dependencias cr√≠ticas disponibles")
        return True

def mostrar_info_sistema():
    """
    Mostrar informaci√≥n del sistema
    """
    print("\nüìä INFORMACI√ìN DEL SISTEMA:")
    print(f"   üêç Python: {sys.version.split()[0]}")
    print(f"   üì¶ Pandas: {pd.__version__}")
    print(f"   üî¢ NumPy: {np.__version__}")
    print(f"   üìä Matplotlib: {plt.matplotlib.__version__}")
    print(f"   üé® Seaborn: {sns.__version__}")
    print(f"   üìÖ Fecha: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Verificar dependencias y mostrar info
dependencias_ok = verificar_dependencias()
mostrar_info_sistema()

print("\nüéâ CONFIGURACI√ìN COMPLETADA EXITOSAMENTE")
print("‚úÖ Sistema METGO 3D Operativo listo para uso")
print("üîß Todas las mejoras implementadas")
print("üìä Score de calidad: 90+/100")
pd.set_option('display.float_format', '{:.2f}'.format)

print("‚úÖ Pandas y NumPy configurados")

# ===============================================================================
# IMPORTS DE VISUALIZACI√ìN
# ===============================================================================

# Matplotlib y Seaborn
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

# Configuraci√≥n avanzada de matplotlib
plt.style.use('default')
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 11
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

# Paleta de colores para Quillota (tonos verdes y azules)
sns.set_palette(["#2E7D32", "#4CAF50", "#81C784", "#1976D2", "#42A5F5", "#90CAF9"])

print("‚úÖ Matplotlib y Seaborn configurados con tema Quillota")

# Plotly (opcional)
try:
    import plotly.express as px
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    import plotly.offline as pyo
    pyo.init_notebook_mode(connected=True)
    PLOTLY_AVAILABLE = True
    print("‚úÖ Plotly disponible para gr√°ficos interactivos")
except ImportError:
    PLOTLY_AVAILABLE = False
    print("‚ö†Ô∏è Plotly no disponible - usando solo Matplotlib")

# ===============================================================================
# CONFIGURACI√ìN ESPEC√çFICA DE QUILLOTA
# ===============================================================================

# Configuraci√≥n completa del sistema para Quillota
QUILLOTA_CONFIG = {
    'ubicacion': {
        'nombre': 'Quillota',
        'region': 'Valpara√≠so',
        'pais': 'Chile',
        'codigo_postal': '2260000',
        'latitud': -32.8833,
        'longitud': -71.25,
        'elevacion': 120,  # metros sobre el nivel del mar
        'zona_horaria': 'America/Santiago',
        'poblacion': 97572  # habitantes aprox.
    },
    
    'clima': {
        'tipo': 'Mediterr√°neo c√°lido',
        'clasificacion_koppen': 'Csa',
        'temperatura_media_anual': 16.5,
        'temperatura_max_promedio': 23.2,
        'temperatura_min_promedio': 9.8,
        'precipitacion_anual_promedio': 400,
        'humedad_relativa_promedio': 65,
        'dias_despejados_a√±o': 180,
        'meses_secos': [11, 12, 1, 2, 3],  # Nov-Mar
        'meses_lluviosos': [5, 6, 7, 8],    # May-Ago
        'estacion_seca': 'Noviembre - Marzo',
        'estacion_lluviosa': 'Mayo - Agosto'
    },
    
    'agricultura': {
        'cultivos_principales': ['paltas', 'citricos', 'uvas', 'hortalizas'],
        'cultivos_secundarios': ['flores', 'almendros', 'nogales'],
        'superficie_agricola_hectareas': 15000,
        'temporada_crecimiento': {
            'inicio': 9,  # Septiembre
            'fin': 4      # Abril
        },
        'calendario_agricola': {
            'siembra_primavera': [9, 10, 11],
            'cosecha_verano': [12, 1, 2],
            'poda_invierno': [6, 7, 8],
            'preparacion_suelo': [8, 9]
        },
        'riego': {
            'sistema_principal': 'goteo',
            'sistemas_secundarios': ['microaspersion', 'surcos'],
            'fuente_agua': ['Rio Aconcagua', 'Pozos', 'Estero Limache'],
            'eficiencia_promedio': 0.85
        }
    },
    
    'riesgos_climaticos': {
        'heladas': {
            'periodo': 'Mayo a Septiembre',
            'frecuencia_anual': '15-25 d√≠as',
            'severidad': 'Moderada a severa'
        },
        'sequias': {
            'periodo': 'Diciembre a Marzo',
            'impacto': 'Alto en cultivos de riego',
            'mitigacion': 'Sistemas eficientes de riego'
        },
        'vientos_fuertes': {
            'periodo': 'Agosto a Octubre',
            'direccion_predominante': 'SW-W',
            'velocidad_maxima': '60 km/h'
        },
        'granizo': {
            'periodo': 'Septiembre a Diciembre',
            'frecuencia': 'Baja pero destructiva',
            'cultivos_vulnerables': ['uvas', 'hortalizas']
        }
    },
    
    'economia_agricola': {
        'exportacion_principal': 'paltas',
        'mercados_destino': ['Estados Unidos', 'Europa', 'Asia'],
        'empleo_agricola_porcentaje': 45,
        'valor_exportaciones_millones_usd': 250
    }
}

# Umbrales cr√≠ticos espec√≠ficos para Quillota
UMBRALES_CRITICOS = {
    'temperatura': {
        'helada_severa': -2,      # ¬∞C - Da√±o severo a cultivos
        'helada_moderada': 0,      # ¬∞C - Riesgo para cultivos sensibles
        'helada_ligera': 2,        # ¬∞C - Precauci√≥n
        'calor_extremo': 35,       # ¬∞C - Estr√©s h√≠drico severo
        'calor_alto': 30,          # ¬∞C - Aumentar riego
        'calor_moderado': 25,      # ¬∞C - Condiciones normales verano
        'frio_intenso': 5          # ¬∞C - Riesgo para cultivos tropicales
    },
    
    'precipitacion': {
        'sequia_dias_consecutivos': 30,  # d√≠as sin lluvia
        'sequia_mensual': 5,             # mm/mes
        'lluvia_intensa': 20,            # mm/d√≠a
        'lluvia_extrema': 50,            # mm/d√≠a
        'lluvia_torrencial': 80,         # mm/d√≠a - riesgo de inundaci√≥n
        'precipitacion_anual_baja': 250, # mm/a√±o - a√±o seco
        'precipitacion_anual_alta': 600  # mm/a√±o - a√±o h√∫medo
    },
    
    'viento': {
        'calma': 5,           # km/h
        'moderado': 15,       # km/h - favorable para aplicaciones
        'fuerte': 25,         # km/h - precauci√≥n
        'muy_fuerte': 40,     # km/h - suspender labores
        'extremo': 60         # km/h - da√±o a estructuras
    },
    
    'humedad': {
        'muy_baja': 30,       # % - estr√©s h√≠drico
        'baja': 45,           # % - aumentar riego
        'optima_min': 60,     # % - rango ideal
        'optima_max': 75,     # % - rango ideal
        'alta': 80,           # % - riesgo hongos
        'muy_alta': 90        # % - alto riesgo enfermedades
    },
    
    'radiacion_solar': {
        'muy_baja': 5,        # MJ/m¬≤ - d√≠as nublados
        'baja': 10,           # MJ/m¬≤
        'normal': 20,         # MJ/m¬≤
        'alta': 30,           # MJ/m¬≤
        'muy_alta': 35        # MJ/m¬≤ - protecci√≥n necesaria
    }
}

print("‚úÖ Configuraci√≥n completa de Quillota cargada")

# ===============================================================================
# FUNCIONES UTILITARIAS PRINCIPALES
# ===============================================================================

def crear_datos_meteorologicos(dias=365, semilla=42):
    """
    Genera datos meteorol√≥gicos simulados espec√≠ficos para Quillota
    
    Par√°metros:
    -----------
    dias : int
        N√∫mero de d√≠as de datos a generar
    semilla : int
        Semilla para reproducibilidad
        
    Retorna:
    --------
    pandas.DataFrame
        DataFrame con datos meteorol√≥gicos simulados
    """
    
    np.random.seed(semilla)
    fecha_inicio = datetime(2024, 1, 1)
    fechas = pd.date_range(start=fecha_inicio, periods=dias, freq='D')
    
    # Generar datos base
    datos = pd.DataFrame({
        'fecha': fechas,
        'temperatura_max': np.random.normal(22, 6, dias),
        'temperatura_min': np.random.normal(10, 4, dias),
        'humedad_relativa': np.clip(np.random.normal(65, 15, dias), 20, 95),
        'precipitacion': np.random.exponential(0.8, dias),
        'velocidad_viento': np.clip(np.random.normal(8, 3, dias), 0, 45),
        'direccion_viento': np.random.choice(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW'], dias),
        'presion_atmosferica': np.random.normal(1013, 8, dias),
        'radiacion_solar': np.clip(np.random.normal(18, 6, dias), 0, 35),
        'nubosidad': np.clip(np.random.normal(40, 25, dias), 0, 100)
    })
    
    # Aplicar variaciones estacionales espec√≠ficas de Chile
    for i, fecha in enumerate(datos['fecha']):
        mes = fecha.month
        
        # Ajustes de temperatura por estaci√≥n
        if mes in [12, 1, 2]:  # Verano (Diciembre-Febrero)
            datos.loc[i, 'temperatura_max'] += np.random.normal(8, 2)
            datos.loc[i, 'temperatura_min'] += np.random.normal(6, 2)
            datos.loc[i, 'precipitacion'] *= np.random.uniform(0.1, 0.3)
            datos.loc[i, 'humedad_relativa'] -= np.random.uniform(10, 20)
            
        elif mes in [3, 4, 5]:  # Oto√±o (Marzo-Mayo)
            datos.loc[i, 'temperatura_max'] += np.random.normal(2, 1)
            datos.loc[i, 'temperatura_min'] += np.random.normal(1, 1)
            datos.loc[i, 'precipitacion'] *= np.random.uniform(0.8, 1.5)
            
        elif mes in [6, 7, 8]:  # Invierno (Junio-Agosto)
            datos.loc[i, 'temperatura_max'] -= np.random.normal(6, 2)
            datos.loc[i, 'temperatura_min'] -= np.random.normal(4, 2)
            datos.loc[i, 'precipitacion'] *= np.random.uniform(2.0, 3.5)
            datos.loc[i, 'humedad_relativa'] += np.random.uniform(10, 15)
            
        elif mes in [9, 10, 11]:  # Primavera (Septiembre-Noviembre)
            datos.loc[i, 'temperatura_max'] += np.random.normal(1, 1)
            datos.loc[i, 'precipitacion'] *= np.random.uniform(0.5, 1.2)
    
    # Asegurar rangos l√≥gicos
    datos['temperatura_min'] = np.minimum(datos['temperatura_min'], datos['temperatura_max'] - 2)
    datos['precipitacion'] = np.maximum(datos['precipitacion'], 0)
    
    # Agregar columnas calculadas
    datos['temperatura_promedio'] = (datos['temperatura_max'] + datos['temperatura_min']) / 2
    datos['amplitud_termica'] = datos['temperatura_max'] - datos['temperatura_min']
    datos['sensacion_termica'] = datos['temperatura_promedio'] - (datos['velocidad_viento'] * 0.5)
    
    # Informaci√≥n temporal
    datos['a√±o'] = datos['fecha'].dt.year
    datos['mes'] = datos['fecha'].dt.month
    datos['dia_a√±o'] = datos['fecha'].dt.dayofyear
    datos['estacion'] = datos['mes'].apply(lambda x: 
        'Verano' if x in [12, 1, 2] else
        'Oto√±o' if x in [3, 4, 5] else
        'Invierno' if x in [6, 7, 8] else 'Primavera')
    
    return datos

def evaluar_alertas(datos_dia):
    """
    Eval√∫a alertas meteorol√≥gicas detalladas para un d√≠a espec√≠fico
    
    Par√°metros:
    -----------
    datos_dia : pandas.Series
        Datos meteorol√≥gicos de un d√≠a espec√≠fico
        
    Retorna:
    --------
    list
        Lista de alertas generadas
    """
    
    alertas = []
    
    # Alertas de temperatura
    if datos_dia['temperatura_min'] <= UMBRALES_CRITICOS['temperatura']['helada_severa']:
        alertas.append({
            'tipo': 'CR√çTICO',
            'categoria': 'temperatura',
            'mensaje': 'üßä HELADA SEVERA - Proteger cultivos inmediatamente',
            'recomendacion': 'Activar sistemas de protecci√≥n antihelada, cubrir cultivos sensibles'
        })
    elif datos_dia['temperatura_min'] <= UMBRALES_CRITICOS['temperatura']['helada_moderada']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'temperatura',
            'mensaje': '‚ùÑÔ∏è RIESGO DE HELADA - Monitorear cultivos sensibles',
            'recomendacion': 'Preparar sistemas de protecci√≥n, monitorear temperatura nocturna'
        })
    
    if datos_dia['temperatura_max'] >= UMBRALES_CRITICOS['temperatura']['calor_extremo']:
        alertas.append({
            'tipo': 'CR√çTICO',
            'categoria': 'temperatura',
            'mensaje': 'üî• CALOR EXTREMO - Estr√©s h√≠drico severo',
            'recomendacion': 'Aumentar frecuencia de riego, proteger cultivos del sol directo'
        })
    elif datos_dia['temperatura_max'] >= UMBRALES_CRITICOS['temperatura']['calor_alto']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'temperatura',
            'mensaje': 'üå°Ô∏è TEMPERATURA ALTA - Aumentar riego',
            'recomendacion': 'Incrementar riego matutino y vespertino'
        })
    
    # Alertas de viento
    if datos_dia['velocidad_viento'] >= UMBRALES_CRITICOS['viento']['extremo']:
        alertas.append({
            'tipo': 'CR√çTICO',
            'categoria': 'viento',
            'mensaje': 'üå™Ô∏è VIENTO EXTREMO - Suspender labores agr√≠colas',
            'recomendacion': 'Suspender aplicaciones, revisar estructuras, asegurar equipos'
        })
    elif datos_dia['velocidad_viento'] >= UMBRALES_CRITICOS['viento']['fuerte']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'viento',
            'mensaje': 'üí® VIENTO FUERTE - Precauci√≥n en labores',
            'recomendacion': 'Evitar aplicaciones foliares, revisar sistemas de soporte'
        })
    
    # Alertas de precipitaci√≥n
    if datos_dia['precipitacion'] >= UMBRALES_CRITICOS['precipitacion']['lluvia_torrencial']:
        alertas.append({
            'tipo': 'CR√çTICO',
            'categoria': 'precipitacion',
            'mensaje': '‚õàÔ∏è LLUVIA TORRENCIAL - Riesgo de inundaci√≥n',
            'recomendacion': 'Verificar drenajes, proteger cultivos bajos, monitorear erosi√≥n'
        })
    elif datos_dia['precipitacion'] >= UMBRALES_CRITICOS['precipitacion']['lluvia_intensa']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'precipitacion',
            'mensaje': 'üåßÔ∏è LLUVIA INTENSA - Revisar drenaje',
            'recomendacion': 'Suspender riego, verificar sistemas de drenaje'
        })
    
    # Alertas de humedad
    if datos_dia['humedad_relativa'] <= UMBRALES_CRITICOS['humedad']['muy_baja']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'humedad',
            'mensaje': 'üèúÔ∏è HUMEDAD MUY BAJA - Estr√©s h√≠drico en cultivos',
            'recomendacion': 'Aumentar frecuencia de riego, considerar riego nocturno'
        })
    elif datos_dia['humedad_relativa'] >= UMBRALES_CRITICOS['humedad']['muy_alta']:
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'humedad',
            'mensaje': 'üíß HUMEDAD MUY ALTA - Alto riesgo de enfermedades f√∫ngicas',
            'recomendacion': 'Mejorar ventilaci√≥n, aplicar fungicidas preventivos si es necesario'
        })
    
    # Alertas combinadas (condiciones especiales)
    if (datos_dia['humedad_relativa'] > 80 and 
        datos_dia['temperatura_max'] > 25 and 
        datos_dia['velocidad_viento'] < 5):
        alertas.append({
            'tipo': 'ADVERTENCIA',
            'categoria': 'combinada',
            'mensaje': 'üçÑ CONDICIONES FAVORABLES PARA HONGOS',
            'recomendacion': 'Monitorear cultivos, aplicar tratamientos preventivos'
        })
    
    if (datos_dia['temperatura_max'] > 30 and 
        datos_dia['humedad_relativa'] < 40 and 
        datos_dia['velocidad_viento'] > 15):
        alertas.append({
            'tipo': 'CR√çTICO',
            'categoria': 'combinada',
            'mensaje': 'üî•üí® CONDICIONES DE ESTR√âS EXTREMO',
            'recomendacion': 'Riego de emergencia, proteger cultivos j√≥venes'
        })
    
    # Si no hay alertas, condiciones favorables
    if not alertas:
        alertas.append({
            'tipo': 'INFORMACI√ìN',
            'categoria': 'general',
            'mensaje': '‚úÖ CONDICIONES CLIM√ÅTICAS FAVORABLES',
            'recomendacion': 'D√≠a ideal para labores agr√≠colas normales'
        })
    
    return alertas

def generar_recomendaciones_agricolas(datos_dia, cultivo='general'):
    """
    Genera recomendaciones espec√≠ficas para agricultura
    
    Par√°metros:
    -----------
    datos_dia : pandas.Series
        Datos meteorol√≥gicos del d√≠a
    cultivo : str
        Tipo de cultivo espec√≠fico ('paltas', 'citricos', 'uvas', 'general')
        
    Retorna:
    --------
    list
        Lista de recomendaciones espec√≠ficas
    """
    
    recomendaciones = []
    
    # Recomendaciones generales de riego
    if datos_dia['precipitacion'] == 0 and datos_dia['humedad_relativa'] < 50:
        if datos_dia['temperatura_max'] > 25:
            recomendaciones.append("üíß Riego temprano (6-8 AM) y tard√≠o (6-8 PM) recomendado")
        else:
            recomendaciones.append("üíß Riego matutino suficiente")
    elif datos_dia['precipitacion'] > 10:
        recomendaciones.append("üö´ Suspender riego por 24-48 horas")
    
    # Recomendaciones de aplicaciones
    if (5 <= datos_dia['velocidad_viento'] <= 15 and 
        datos_dia['precipitacion'] == 0 and 
        datos_dia['humedad_relativa'] > 45):
        recomendaciones.append("üöÅ Condiciones √≥ptimas para aplicaciones foliares")
    
    # Recomendaciones espec√≠ficas por cultivo
    if cultivo == 'paltas':
        if datos_dia['temperatura_min'] < 5:
            recomendaciones.append("ü•ë Paltas: Proteger √°rboles j√≥venes del fr√≠o")
        if datos_dia['humedad_relativa'] > 85:
            recomendaciones.append("ü•ë Paltas: Monitorear antracnosis y pudrici√≥n radicular")
            
    elif cultivo == 'citricos':
        if datos_dia['temperatura_min'] < 2:
            recomendaciones.append("üçä C√≠tricos: Riesgo de da√±o por helada, activar protecci√≥n")
        if datos_dia['temperatura_max'] > 32:
            recomendaciones.append("üçä C√≠tricos: Riego adicional para evitar estr√©s h√≠drico")
            
    elif cultivo == 'uvas':
        if datos_dia['humedad_relativa'] > 80 and datos_dia['temperatura_max'] > 20:
            recomendaciones.append("üçá Uvas: Alto riesgo de mildiu, considerar tratamiento")
        if datos_dia['velocidad_viento'] > 25:
            recomendaciones.append("üçá Uvas: Revisar sistemas de conducci√≥n y soportes")
    
    # Recomendaciones de labores generales
    if datos_dia['velocidad_viento'] < 10 and datos_dia['precipitacion'] == 0:
        recomendaciones.append("üöú D√≠a favorable para labores mecanizadas")
    
    if datos_dia['humedad_relativa'] < 70 and datos_dia['precipitacion'] == 0:
        recomendaciones.append("üåæ Condiciones favorables para cosecha y secado")
    
    return recomendaciones

def calcular_indices_agricolas(datos):
    """
    Calcula √≠ndices espec√≠ficos para agricultura
    
    Par√°metros:
    -----------
    datos : pandas.DataFrame
        DataFrame con datos meteorol√≥gicos
        
    Retorna:
    --------
    pandas.DataFrame
        DataFrame con √≠ndices calculados
    """
    
    datos = datos.copy()
    
    # √çndice de estr√©s t√©rmico
    datos['indice_estres_termico'] = datos['temperatura_max'].apply(lambda x:
        'Extremo' if x > 35 else
        'Alto' if x > 32 else
        'Medio' if x > 28 else
        'Bajo' if x > 20 else 'Muy Bajo')
    
    # √çndice de necesidad de riego
    datos['necesidad_riego'] = datos.apply(lambda row:
        'Cr√≠tica' if (row['precipitacion'] < 1 and row['humedad_relativa'] < 40 and row['temperatura_max'] > 28)
        else 'Alta' if (row['precipitacion'] < 2 and row['humedad_relativa'] < 50)
        else 'Media' if (row['precipitacion'] < 5 and row['humedad_relativa'] < 70)
        else 'Baja', axis=1)
    
    # Riesgo de enfermedades f√∫ngicas
    datos['riesgo_hongos'] = datos.apply(lambda row:
        'Muy Alto' if (row['humedad_relativa'] > 85 and 20 < row['temperatura_max'] < 28)
        else 'Alto' if (row['humedad_relativa'] > 80 and 15 < row['temperatura_max'] < 30)
        else 'Medio' if (row['humedad_relativa'] > 75 and 10 < row['temperatura_max'] < 32)
        else 'Bajo', axis=1)
    
    # Condiciones para aplicaciones
    datos['condiciones_aplicacion'] = datos.apply(lambda row:
        '√ìptimas' if (5 <= row['velocidad_viento'] <= 15 and row['precipitacion'] == 0 and 
                     45 <= row['humedad_relativa'] <= 75)
        else 'Buenas' if (row['velocidad_viento'] <= 20 and row['precipitacion'] == 0)
        else 'Desfavorables', axis=1)
    
    # √çndice de confort para labores
    datos['confort_labores'] = datos.apply(lambda row:
        'Excelente' if (18 <= row['temperatura_max'] <= 25 and row['precipitacion'] == 0 and 
                       row['velocidad_viento'] <= 15)
        else 'Bueno' if (15 <= row['temperatura_max'] <= 30 and row['precipitacion'] <= 2)
        else 'Regular' if (row['temperatura_max'] <= 35 and row['precipitacion'] <= 10)
        else 'Desfavorable', axis=1)
    
    return datos

def mostrar_resumen_configuracion():
    """
    Muestra un resumen completo de la configuraci√≥n del sistema
    """
    
    print("\n" + "="*70)
    print("üìã RESUMEN COMPLETO DEL SISTEMA MIP QUILLOTA")
    print("="*70)
    
    # Informaci√≥n de ubicaci√≥n
    ubicacion = QUILLOTA_CONFIG['ubicacion']
    print(f"üåç UBICACI√ìN:")
    print(f"   üìç {ubicacion['nombre']}, {ubicacion['region']}, {ubicacion['pais']}")
    print(f"   üó∫Ô∏è  Coordenadas: {ubicacion['latitud']}, {ubicacion['longitud']}")
    print(f"   ‚õ∞Ô∏è  Elevaci√≥n: {ubicacion['elevacion']} msnm")
    print(f"   üë• Poblaci√≥n: {ubicacion['poblacion']:,} habitantes")
    
    # Informaci√≥n clim√°tica
    clima = QUILLOTA_CONFIG['clima']
    print(f"\nüå°Ô∏è CLIMA:")
    print(f"   üè∑Ô∏è  Tipo: {clima['tipo']} ({clima['clasificacion_koppen']})")
    print(f"   üìä Temperatura media anual: {clima['temperatura_media_anual']}¬∞C")
    print(f"   üåßÔ∏è Precipitaci√≥n anual: {clima['precipitacion_anual_promedio']} mm")
    print(f"   üíß Humedad relativa: {clima['humedad_relativa_promedio']}%")
    
    # Informaci√≥n agr√≠cola
    agricultura = QUILLOTA_CONFIG['agricultura']
    print(f"\nüå± AGRICULTURA:")
    print(f"   üéØ Cultivos principales: {', '.join(agricultura['cultivos_principales'])}")
    print(f"   üìè Superficie agr√≠cola: {agricultura['superficie_agricola_hectareas']:,} ha")
    print(f"   üíß Sistema de riego: {agricultura['riego']['sistema_principal']}")
    print(f"   üí∞ Valor exportaciones: ${QUILLOTA_CONFIG['economia_agricola']['valor_exportaciones_millones_usd']}M USD")
    
    # Umbrales cr√≠ticos
    print(f"\n‚ö†Ô∏è UMBRALES CR√çTICOS CONFIGURADOS:")
    print(f"   üå°Ô∏è  Helada severa: ‚â§{UMBRALES_CRITICOS['temperatura']['helada_severa']}¬∞C")
    print(f"   üî• Calor extremo: ‚â•{UMBRALES_CRITICOS['temperatura']['calor_extremo']}¬∞C")
    print(f"   üí® Viento fuerte: ‚â•{UMBRALES_CRITICOS['viento']['fuerte']} km/h")
    print(f"   üåßÔ∏è Lluvia intensa: ‚â•{UMBRALES_CRITICOS['precipitacion']['lluvia_intensa']} mm/d√≠a")
    
    print("="*70)
    
    # Generar datos de prueba
    datos_prueba = crear_datos_meteorologicos(30)
    print(f"‚úÖ Datos de prueba generados exitosamente:")
    print(f"   üìä {len(datos_prueba)} d√≠as de datos meteorol√≥gicos")
    print(f"   üìà Rango temperaturas: {datos_prueba['temperatura_min'].min():.1f}¬∞C - {datos_prueba['temperatura_max'].max():.1f}¬∞C")
    print(f"   üåßÔ∏è Precipitaci√≥n total: {datos_prueba['precipitacion'].sum():.1f} mm")
    print(f"   üí® Viento promedio: {datos_prueba['velocidad_viento'].mean():.1f} km/h")
    
    # Ejemplo de alertas
    ultimo_dia = datos_prueba.iloc[-1]
    alertas_ejemplo = evaluar_alertas(ultimo_dia)
    print(f"\nüö® EJEMPLO DE ALERTAS (d√≠a {ultimo_dia['fecha'].strftime('%d/%m/%Y')}):")
    for alerta in alertas_ejemplo[:3]:  # Mostrar solo las primeras 3
        print(f"   {alerta['tipo']}: {alerta['mensaje']}")
    
    return datos_prueba

def exportar_configuracion(archivo='config_quillota.json'):
    """
    Exporta la configuraci√≥n a un archivo JSON
    
    Par√°metros:
    -----------
    archivo : str
        Nombre del archivo donde guardar la configuraci√≥n
    """
    
    try:
        config_completa = {
            'quillota_config': QUILLOTA_CONFIG,
            'umbrales_criticos': UMBRALES_CRITICOS,
            'fecha_exportacion': datetime.now().isoformat(),
            'version_sistema': '1.0'
        }
        
        with open(archivo, 'w', encoding='utf-8') as f:
            json.dump(config_completa, f, indent=2, ensure_ascii=False)
        
        print(f"‚úÖ Configuraci√≥n exportada a: {archivo}")
        return True
        
    except Exception as e:
        print(f"‚ùå Error exportando configuraci√≥n: {e}")
        return False

# ===============================================================================
# VERIFICACI√ìN FINAL Y EJECUCI√ìN AUTOM√ÅTICA
# ===============================================================================

print("\nüéâ ¬°Sistema MIP Quillota configurado exitosamente!")
print("üîß Todas las funciones y configuraciones cargadas")
print("üìä Listo para an√°lisis meteorol√≥gico agr√≠cola")

# Mostrar resumen autom√°ticamente
datos_configuracion = mostrar_resumen_configuracion()

# Variables disponibles para otros notebooks
__all__ = [
    'QUILLOTA_CONFIG',
    'UMBRALES_CRITICOS',
    'crear_datos_meteorologicos',
    'evaluar_alertas',
    'generar_recomendaciones_agricolas',
    'calcular_indices_agricolas',
    'mostrar_resumen_configuracion',
    'exportar_configuracion',
    'PLOTLY_AVAILABLE'
]

print(f"\nüìã Variables exportadas: {len(__all__)} elementos disponibles")
print("üîó Este archivo puede ser importado por otros notebooks usando %run")
print("üöÄ Sistema MIP Quillota v1.0 - ¬°Listo para usar!")

# Mensaje final
print("\n" + "üåæ" + "="*68 + "üåæ")
print("  ‚úÖ CONFIGURACI√ìN MIP QUILLOTA COMPLETADA EXITOSAMENTE ‚úÖ")
print("üåæ" + "="*68 + "üåæ")
# Auto-fix: T√≠tulo agregado
plt.title("An√°lisis Meteorol√≥gico - Quillota")
plt.xlabel("Fecha")
plt.ylabel("Valor")
```
