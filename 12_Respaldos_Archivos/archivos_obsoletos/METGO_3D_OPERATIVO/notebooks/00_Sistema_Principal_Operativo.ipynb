```python
# =============================================================================
# INICIALIZACIÃ“N DEL SISTEMA
# =============================================================================

# Inicializar sistema meteorolÃ³gico
sistema = SistemaMeteorologicoQuillota()

print("\nğŸš€ Sistema MeteorolÃ³gico Quillota inicializado")
print(f"ğŸ“ UbicaciÃ³n: {sistema.config['QUILLOTA']['nombre']}, {sistema.config['QUILLOTA']['region']}")
print(f"ğŸ—ºï¸ Coordenadas: {sistema.config['QUILLOTA']['coordenadas']['latitud']}, {sistema.config['QUILLOTA']['coordenadas']['longitud']}")
print("\nâœ… Todos los mÃ³dulos cargados exitosamente:")
print("   - API MeteorolÃ³gica (OpenMeteo)")
print("   - Validador de Datos")
print("   - Pipeline de Machine Learning")
print("   - Dashboard de VisualizaciÃ³n")
print("   - Sistema de Logging")
```

```python
# =============================================================================
# CARGA DE DATOS METEOROLÃ“GICOS
# =============================================================================

print("\nğŸ“Š CARGANDO DATOS METEOROLÃ“GICOS")
print("=" * 50)

# Cargar datos meteorolÃ³gicos (30 dÃ­as)
datos = sistema.cargar_datos_meteorologicos(dias=30, fuente="auto")

print(f"\nâœ… Datos cargados exitosamente:")
print(f"   ğŸ“… PerÃ­odo: {datos['fecha'].min().strftime('%d/%m/%Y')} - {datos['fecha'].max().strftime('%d/%m/%Y')}")
print(f"   ğŸ“Š Registros: {len(datos)} dÃ­as")
print(f"   ğŸŒ¡ï¸ Temperatura promedio: {datos['temperatura_max'].mean():.1f}Â°C")
print(f"   ğŸŒ§ï¸ PrecipitaciÃ³n total: {datos['precipitacion'].sum():.1f} mm")
print(f"   ğŸ’§ Humedad promedio: {datos['humedad_relativa'].mean():.0f}%")
print(f"   ğŸ’¨ Viento promedio: {datos['velocidad_viento'].mean():.1f} km/h")

# Mostrar primeras filas
print("\nğŸ“‹ Primeras 5 filas de datos:")
display(datos.head())
```

```python
# =============================================================================
# ANÃLISIS METEOROLÃ“GICO COMPLETO
# =============================================================================

print("\nğŸŒ¤ï¸ ANÃLISIS METEOROLÃ“GICO COMPLETO")
print("=" * 50)

# Realizar anÃ¡lisis meteorolÃ³gico
analisis = sistema.analizar_datos(datos)

print("\nâœ… AnÃ¡lisis completado exitosamente:")
print(f"   ğŸ“Š Resumen general: {analisis['resumen_general']['total_dias']} dÃ­as analizados")
print(f"   ğŸŒ¡ï¸ Temperatura mÃ¡xima: {analisis['estadisticas_temperatura']['temperatura_maxima']:.1f}Â°C")
print(f"   ğŸŒ¡ï¸ Temperatura mÃ­nima: {analisis['estadisticas_temperatura']['temperatura_minima']:.1f}Â°C")
print(f"   ğŸ§Š DÃ­as con heladas: {analisis['estadisticas_temperatura']['dias_helada']}")
print(f"   ğŸ”¥ DÃ­as calor extremo: {analisis['estadisticas_temperatura']['dias_calor_extremo']}")
print(f"   ğŸŒ§ï¸ PrecipitaciÃ³n total: {analisis['analisis_precipitacion']['precipitacion_total']:.1f} mm")
print(f"   â˜” DÃ­as con lluvia: {analisis['analisis_precipitacion']['dias_con_lluvia']}")
print(f"   ğŸŒ± Grados-dÃ­a totales: {analisis['indices_agricolas']['grados_dia_total']:.1f}")
print(f"   ğŸ’§ Confort tÃ©rmico: {analisis['indices_agricolas']['confort_termico']}")
print(f"   ğŸ’¦ Necesidad riego: {analisis['indices_agricolas']['necesidad_riego']}")
print(f"   ğŸ„ Riesgo hongos: {analisis['indices_agricolas']['riesgo_hongos']}")

# Mostrar anÃ¡lisis detallado
print("\nğŸ“‹ AnÃ¡lisis detallado:")
for categoria, datos_categoria in analisis.items():
    if isinstance(datos_categoria, dict):
        print(f"\n{categoria.upper()}:")
        for clave, valor in datos_categoria.items():
            if isinstance(valor, (int, float)):
                print(f"   {clave}: {valor:.2f}")
            else:
                print(f"   {clave}: {valor}")
```

```python
# =============================================================================
# EVALUACIÃ“N DE ALERTAS METEOROLÃ“GICAS
# =============================================================================

print("\nğŸš¨ EVALUACIÃ“N DE ALERTAS METEOROLÃ“GICAS")
print("=" * 50)

# Evaluar alertas
alertas = sistema.evaluar_alertas(datos)

print(f"\nğŸ“Š Alertas generadas: {len(alertas)}")

if alertas:
    print("\nğŸš¨ ALERTAS ACTIVAS:")
    for i, alerta in enumerate(alertas, 1):
        print(f"\n{i}. {alerta['mensaje']}")
        print(f"   ğŸ“… Fecha: {alerta['fecha']}")
        print(f"   âš ï¸ Severidad: {alerta['severidad'].upper()}")
        print(f"   ğŸ“Š Valor: {alerta['valor']:.1f}")
        print(f"   ğŸ·ï¸ Tipo: {alerta['tipo']}")
else:
    print("\nâœ… No se generaron alertas meteorolÃ³gicas")
    print("   Las condiciones meteorolÃ³gicas estÃ¡n dentro de rangos normales")

# Crear grÃ¡fico de alertas
if alertas:
    sistema.dashboard.crear_grafico_alertas(alertas)
```

```python
# =============================================================================
# ENTRENAMIENTO DE MODELOS DE MACHINE LEARNING
# =============================================================================

print("\nğŸ¤– ENTRENAMIENTO DE MODELOS DE MACHINE LEARNING")
print("=" * 50)

# Entrenar modelos ML
resultados_ml = sistema.entrenar_modelos_ml(datos)

print("\nâœ… Entrenamiento de modelos completado:")

for variable, resultado in resultados_ml.items():
    print(f"\nğŸ“Š {variable.upper()}:")
    print(f"   ğŸ† Mejor modelo: {resultado['mejor_modelo']}")
    print(f"   ğŸ“ˆ Mejor score CV: {resultado['mejor_score']:.4f}")
    print(f"   ğŸ“Š Total modelos: {resultado['metricas_generales']['total_modelos']}")
    print(f"   ğŸ¯ Mejor RÂ²: {resultado['metricas_generales']['mejor_r2_test']:.4f}")
    print(f"   ğŸ“‰ Mejor RMSE: {resultado['metricas_generales']['mejor_rmse_test']:.4f}")
    
    print(f"\n   ğŸ“‹ MÃ‰TRICAS POR MODELO:")
    for nombre_modelo, info_modelo in resultado['modelos'].items():
        print(f"   {nombre_modelo}:")
        print(f"     RÂ² Test: {info_modelo['metricas']['test_r2']:.4f}")
        print(f"     RMSE Test: {info_modelo['metricas']['test_rmse']:.4f}")
        print(f"     MAE Test: {info_modelo['metricas']['test_mae']:.4f}")
        print(f"     CV Score: {info_modelo['cv_mean']:.4f} Â± {info_modelo['cv_std']:.4f}")
```

```python
# =============================================================================
# CREACIÃ“N DE DASHBOARD METEOROLÃ“GICO
# =============================================================================

print("\nğŸ“Š CREACIÃ“N DE DASHBOARD METEOROLÃ“GICO")
print("=" * 50)

# Crear dashboard completo
sistema.crear_dashboard(analisis, datos)

print("\nâœ… Dashboard meteorolÃ³gico creado exitosamente:")
print("   ğŸ“ˆ GrÃ¡fico de temperaturas")
print("   ğŸŒ§ï¸ GrÃ¡fico de precipitaciÃ³n")
print("   ğŸ’§ GrÃ¡fico de humedad y viento")
print("   ğŸ“Š Dashboard interactivo (HTML)")
print("   ğŸ“‹ GrÃ¡fico de estadÃ­sticas")

# Crear grÃ¡fico de estadÃ­sticas
sistema.dashboard.crear_grafico_estadisticas(analisis)

print("\nğŸ“ Archivos generados:")
print("   - logs/grafico_temperaturas.png")
print("   - logs/grafico_precipitacion.png")
print("   - logs/grafico_humedad_viento.png")
print("   - logs/grafico_estadisticas.png")
print("   - logs/dashboard_interactivo.html")
```

```python
# =============================================================================
# GENERACIÃ“N DE REPORTE EJECUTIVO
# =============================================================================

print("\nğŸ“‹ GENERACIÃ“N DE REPORTE EJECUTIVO")
print("=" * 50)

# Generar reporte completo
reporte_path = sistema.generar_reporte(analisis, alertas)

print(f"\nâœ… Reporte ejecutivo generado:")
print(f"   ğŸ“„ Archivo: {reporte_path}")
print(f"   ğŸ“Š Datos procesados: {len(datos)} registros")
print(f"   ğŸš¨ Alertas generadas: {len(alertas)}")
print(f"   ğŸ¤– Modelos entrenados: {len(resultados_ml)}")
print(f"   ğŸ“ˆ GrÃ¡ficos creados: 5")

# Mostrar resumen del reporte
print("\nğŸ“‹ RESUMEN DEL REPORTE:")
print(f"   ğŸ“… Fecha de anÃ¡lisis: {analisis['fecha_analisis']}")
print(f"   ğŸ“Š PerÃ­odo analizado: {analisis['resumen_general']['total_dias']} dÃ­as")
print(f"   ğŸŒ¡ï¸ Temperatura promedio: {analisis['resumen_general']['temperatura_promedio']:.1f}Â°C")
print(f"   ğŸŒ§ï¸ PrecipitaciÃ³n total: {analisis['resumen_general']['precipitacion_total']:.1f} mm")
print(f"   â˜” DÃ­as con lluvia: {analisis['resumen_general']['dias_con_lluvia']}")

if alertas:
    print(f"\nğŸš¨ ALERTAS EN EL REPORTE:")
    for alerta in alertas:
        print(f"   - {alerta['mensaje']}")
else:
    print("\nâœ… No hay alertas meteorolÃ³gicas en el perÃ­odo analizado")
```
